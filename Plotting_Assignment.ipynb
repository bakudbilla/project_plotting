```python
plt.hist(data["co2"], bins=10, color='blue')
plt.title("Distribution of CO2")
plt.savefig("histogram_co2")
plt.xlabel("CO2")
plt.ylabel("frequency")
plt.show()
```

## INTERPRETATION:

This histogram distribution for co2 shows a  right skewed distribution,indicating that most vehicles emit lower amount of co2(230-580) with fewer vehicles emiting higer levels.From this histograms; it implies most vehicles are designed to be decrease emissions of co2 and to fight climate change.

```python
plt.hist(data["ghgScore"], bins=4, color='black', edgecolor="pink")
plt.title("Distribution of ghgScore")
plt.savefig("histogram_ghgScore")
plt.xlabel("ghgScore")
plt.ylabel("frequency")
plt.show()
```

# INTERPRETATION:
The distribution is relatively uniform with a slight concentration in the middle(4-6).implying that most vehicles have an average greenhouse gas score, reflecting a balance between high and low environmental impact vehicles.The ghgscore is used to measure vehicles environmental impact, High ghgScore indicates high impact on the environment in relation to emissions by vehicles.

```python
plt.hist(data["cylinders"], bins=10, color='brown')
plt.title("Distribution of Cylinders")
plt.savefig("histogram_cylinders")
plt.xlabel("cylinders")
plt.ylabel("frequency")
plt.show()
```

## **INTERPRETATION**:
 This histogram is left-skewed with many vehicles having small number of cylinders(4-6).Less cylinders implies less fuel comsumptions which implies the vehicles in this dataset are economical(not much money needed to fuel cars).

```python
numeric_data = data.select_dtypes(include=['int64', 'float64'])

# Create a heatmap to visualize the correlation between the numeric columns
plt.figure(figsize=(8, 6))
correlation_matrix = numeric_data.corr()
sns.heatmap(correlation_matrix, annot=False, cmap='coolwarm')
plt.savefig("heatmap.png")
plt.title('FUEL DATA Heatmap')
plt.show()
```

# INTERPRETATION:
The heatmap shows strong correlations between most variables in the dataset.Features such as city, UCity, highway, UHighway, and comb have strong positive correlations with each other, which implies they are directly related in measuring fuel effienciency.The co2 emissions have a strong negative correlation with feScore and ghgScore,implying less environmental pollutions as higher emissions of co2 leads improves environmental conditions.
